```python
## Importamos Librerias 
import random as rm

## Objetos que nos ayudaran a crear las matrices.
class Secuencias():

    def __init__(self):
        self.arreglo = []

    def agregar_secuencia(self,sec):
        variable = []
        for i in range(0,len(sec),1):
            variable.append(sec[i])

        self.arreglo.append(variable)

    def imprimir_arreglo (self):
        for row  in self.arreglo:
            print(row)
```

```python
## Declaración de Matrices y secuencias iniciales con las que se iniciara el universo.

# nm = número de matrices con las que vamos a iniciar
# tm ) número de secuencias por matriz

nm = int(input("Ingrese el número de matrices que desea ingresar: "))
tm = int(input("Ingrese el número de secuencias por matriz: "))

matrices = []


for i in range (nm):
    secuencias = []
    for j in range (tm):
        sec = input(f"Ingrese la secuencia {j+1} de la matriz {i+1}: ")
        secuencias.append(sec)
    matrices.append(secuencias)

for row in matrices:
    print (row)
```

```python
## Crear universo y poblarlo

# repc = cantidad de veces que desea reproducir cada matriz

repc  = int(input(f"Ingrese la cantidad de veces que desea reproducir cada matriz de secuencias: "))

universo  = []
for k in range (len(matrices)):
    for j in range (repc):
        objeto = Secuencias()
        for i in range (len(matrices[k])):
            objeto.agregar_secuencia(matrices[k][i])
        universo.append(objeto.arreglo)

for row in universo:
    print (row)
```

```python
## Encontrar el elemento más grande de las matrices
x = 0

for i in range(len(matrices)):
    for j in range(len(matrices[i])):
        if len(matrices[i][j])> x:
            x = len(matrices[i][j])

print (x)

```

```python
## Creamos una funcion reutilizable para agregar Gaps para rellenar matrices

def Agregar_Gap(mat,n):
    for i in range (len(mat)):
        while len(mat[i]) < n:
            posicion = rm.randint(0,len(mat[i]))
            mat[i].insert(posicion, "-")
```

```python
## Agregamos gaps
## nga = numero de gapas adicionales al elemento más grande

nga = int(input("Ingrege el número de gaps adicionales que desea agregar: "))

for i in range (len(universo)):
    Agregar_Gap(universo[i],nga+x)

for row in universo[1]:
    print(row)
```

```python
## Funcion para evaluacion de las matrices de la matriz

def calificar (mat):
    cal = 0
    for i in range (nga+x):
        temp = []
        for j in range (tm):
            temp.append(mat[j][i])
        for k in range (len(temp)):
            if temp[k] != "-":
                cal +=  (temp.count(temp[k])-1)
    return cal 

    
```

```python
## Evaluar universo de matrices

calificaciones = []
for i in range (len(universo)):
    calificaciones.append(calificar(universo[i]))


for i in range (len(universo)):
    print(f"Matriz {i+1}:")
    print(f"calificacion: {calificaciones[i]}:")
    for row in universo[i]:
        print(row)
```

```python
def cruza (p1, p2):
    hijo1 = []
    hijo2 = []
    for i in range (len(p1)):
        sec1 = p1[i]
        sec2 = p2[i]
        cruza = hijos (sec1,sec2)
        hijo1.append(cruza[0])
        hijo2.append(cruza[1])

    return (hijo1,hijo2)
   

```
